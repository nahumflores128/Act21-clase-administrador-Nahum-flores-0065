```python
class Administrador:
    def __init__(self, nombre, telefono, edad):
        # Atributos: Datos del administrador
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Planificar
    def planificar(self):
        print(f"[PLANIFICACIÓN]: {self.nombre} está elaborando el plan estratégico y asignando recursos.")

    # Función 2: Supervisar
    def supervisar(self):
        print(f"[SUPERVISIÓN]: {self.nombre} está revisando el cumplimiento de las metas del equipo.")

    # Función 3: Reportar
    def reportar(self):
        print(f"[REPORTE]: El administrador {self.nombre} (Edad: {self.edad}) está enviando el informe mensual a la gerencia.")

# --- Creación de Objetos (Instancias) ---

# Objeto 1: administrador1 con datos de ejemplo
administrador1 = Administrador("Roberto Méndez", "555-123-4567", 45)

# Objeto 2: administrador2 con datos de ejemplo
administrador2 = Administrador("Laura Vizcarra", "555-987-6543", 32)

# --- Ejecución y Mensajes ---

print(f"--- Ficha del Administrador 1 ---")
print(f"Nombre: {administrador1.nombre} | Teléfono: {administrador1.telefono}")
administrador1.planificar()
administrador1.supervisar()

print("\n" + "-"*45 + "\n")

print(f"--- Ficha del Administrador 2 ---")
print(f"Nombre: {administrador2.nombre} | Edad: {administrador2.edad}")
administrador2.reportar()
administrador2.planificar()

print("Progama creado por Nahum NL = 18")
```
